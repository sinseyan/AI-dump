```python
import pandas as pd
import matplotlib.pyplot as plt
#from sklearn import cluster
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('wine-clustering.csv')
data.head()
```

```python
kmeans = KMeans(3)
kmeans.fit(data[['Alcohol','OD280']])
labels = kmeans.labels_
print(labels)
plt.scatter(data['Alcohol'], data['OD280'], c = labels)
plt.show()
```

```python
data['cluster'] = labels
data.sort_values(by='cluster')
```

```python
kmeans = KMeans(3)
kmeans.fit(data[['Alcohol', 'OD280']])
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
plt.scatter(data['Alcohol'], data['Proline'], c=labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='x', c='black')
plt.show()
```

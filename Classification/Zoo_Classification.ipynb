```python
import pandas as pd
import numpy

#Packages for Classification algorithm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
zoo = pd.read_csv('zoo.csv')
zoo.head()
```

```python
numitems = len(zoo)
percenttrain = 0.85
numtrain = int(numitems*percenttrain)
numtest = numitems - numtrain
print('Total Records', numitems)
print ('Training set', numtrain, 'items')
print('Test set', numtest, 'items')
zooTrain = zoo[0:numtrain]
zooTest = zoo[numtrain:]
```

```python
cats = []
for i in range(len(zoo)):
    if zoo.loc[i]['class_type'] == 1:
        cats.append('Mammal')
    elif zoo.loc[i]['class_type'] == 2:
        cats.append('Bird')
    elif zoo.loc[i]['class_type'] == 3:
        cats.append('Reptile')
    elif zoo.loc[i]['class_type'] == 4:
        cats.append('Fish')
    elif zoo.loc[i]['class_type'] == 5:
        cats.append('Amphibian')
    elif zoo.loc[i]['class_type'] == 6:
        cats.append('Bug')
    elif zoo.loc[i]['class_type'] == 7:
        cats.append('Invertebrae')
    else:
        cats.append('Unknown')

animalType = ['Mammal','Bird','Reptile','Fish','Amphibian','Bug','Invertebrae','Unknown']
zoo['category'] = cats
zoo.head()

        
```

```python
features = ['hair','feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','breathes','venomous','fins','legs','tail','domestic','catsize']
neighbors = 3
classifier = KNeighborsClassifier(neighbors)
#Build KNN Model
classifier.fit(zooTrain[features], zooTrain['class_type'])
#Predict/Classify
predictions = classifier.predict(zooTest[features])
# Calculate accuracy
numtrain = len(zooTrain)
numtest = len(zooTest)
correct = 0
for i in range(numtest):
  print (zooTest.loc[numtrain+i]['animal_name'],'Predicted:', animalType[predictions[i]-1], ' Actual:', animalType[zooTest.
loc[numtrain+i]['class_type']-1])
  if predictions[i] == zooTest.loc[numtrain+i]['class_type']: correct +=1
print('Correct prediction', correct, 'out of', numtest)
print ('Accuracy:', float(correct)/float(numtest))
```
